```python
# reference https://www.oneoffcoder.com/2019/10/02/latent-dirichlet-allocation/
# https://towardsdatascience.com/evaluate-topic-model-in-python-latent-dirichlet-allocation-lda-7d57484bb5d0

from gensim import corpora
from gensim.models import LdaModel
from gensim.models.coherencemodel import CoherenceModel
```

```python
def topic_coherence_lda(corpus, dictionary, k):
    
    lda = LdaModel(corpus, id2word=dictionary, num_topics=k, 
                   random_state=19, iterations=50, passes=2, per_word_topics = False)
    cm = CoherenceModel(model=lda, corpus=corpus, coherence='u_mass')
    coherence = cm.get_coherence()
    print('{}: {}'.format(k, coherence))
    
    return k, coherence
```

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
#wnl.lemmatize('Koreans')

from nltk.stem.porter import *
stemmer = PorterStemmer()
#stemmer.stem(wnl.lemmatize('Koreans'))

import gensim
from gensim.parsing.preprocessing import STOPWORDS
st_words = gensim.parsing.preprocessing.STOPWORDS

def manual_tokenizer(txt):
    
    L = []
    word_ =""
    
    for char in txt:
        # if character is alphabetical, then add it to the string
        if char.isalpha():
            word_ += char
        
        # st_words is the list of gensim stopwords
        elif (len(word_) > 2) and (word_ not in st_words):
            #WORD = stemmer.stem(wnl.lemmatize(word_))
            WORD = wnl.lemmatize(word_)
            L.append(WORD)
            word_ = ""
        
        else:
            word_ = ""
            
    return L

# Build List of list tokenizer
def dic_tokenizer(dic_):
    
    L = []
    count = 0
    
    for id_ in dic_.keys():     
        content = dic_[id_]
        tokens = manual_tokenizer(content)
        L.append(tokens)
        
    return L
```

```python
# import txt data from the prepared dataset
df = pd.read_csv("training_text.csv")
df = df.iloc[:,1:]

df["content"] = df["content"].map(lambda x: x.lower())

df_dic = {}

for index in df.index:

    id_ = df.at[index,"id"]
    content = df.at[index,"content"]
    df_dic[id_] = content
    
preprocessed_data = dic_tokenizer(df_dic)

dictionary_LDA = corpora.Dictionary(preprocessed_data)

dictionary_LDA.filter_extremes(no_above=0.6)

corpus = [dictionary_LDA.doc2bow(list_of_tokens) for list_of_tokens in preprocessed_data]
```

```python
lda_scores = [topic_coherence_lda(corpus, dictionary_LDA, k) for k in range(10, 50)]
```
